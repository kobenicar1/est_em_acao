# Questão 1
Escreva uma função que verifica se um número é par. Se for, imprima uma mensagem. Se não for, imprima outra mensagem.

```python
def ip():  #um número n é par se o resto da divisão n/2 é 0
    numero = int(input("digite um número: "))
    resto = numero%2
    if resto == 0:
        print('Este é um número par')
    else:
        print('Este é um número impar')
ip()
```

# Questão 2
Escreva uma função que calcule a área de um círculo, dado um valor de raio. Peça um valor de raio ao usuário e calcule o valor da função. (Dica: Use o pacote ``math`` para conseguir o valor de pi)

```python
import math as m # importação de math

def cir(): # a área de um círculo é dada por pi*r². utilizou-se m.pi para a constante pi
    r = float(input('Diga o raio do círculo: '))
    a = m.pi*r**2
    print('A área do círculo desejado é', a)
cir()
```

 (Algoritmo de ordenação: Bubble sort). Bubble sort é um algoritmo muito simples. Ele começa no início da lista e compara os dois primeiros elementos. Se o primeiro for maior do que o segundo, eles são trocados. Esse processo se repete para cada par subsequente, até o final da lista. Depois o algoritmo volta ao início da lista e recomeça o processo, repetindo-o até não haver mais trocas. Implemente uma função que recebe uma lista e retorna seus elementos ordenados usando o algoritmo bubble sort.
 
 Use a seguinte lista para testar ``l = [10, 1, 2, 8, 0, 9, -1, 11, 5, 4]``
 
 (Dica: Você vai precisar de mais de um tipo de loop nesse exercício)

```python
def bs(l):
    tl = len(l)
    for p in range(tl-1):
        for ev in range(tl-1):
            if l[ev]>l[ev+1]:
                l[ev], l[ev+1] = l[ev+1], l[ev]
                print(l)
    print('A lista ordenada é',l)
bs([10, 1, 2, 8, 0, 9, -1, 11, 5, 4])
```
